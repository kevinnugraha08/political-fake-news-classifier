```python
import pandas as pd 
```

```python
df_cnn = pd.read_excel('cnn_indonesia.xlsx')
df_tbh = pd.read_excel('turnbackhoax.xlsx')
```

```python
# --- CEK INFO DATA MENTAH CNN ---
print("--- INFO DATA MENTAH CNN INDONESIA ---")
df_cnn.info()
```

```python
# --- CEK INFO DATA MENTAH TURNBACKHOAX ---
print("\n--- INFO DATA MENTAH TURNBACKHOAX ---")
df_tbh.info()
```

```python
# Menentukan kolom yang akan ditampilkan
kolom_pilihan = ['Timestamp', 'Title', 'FullText', 'Category']

# Mengatur tampilan agar menampilkan total 10 baris (5 atas, 5 bawah) dengan titik-titik di tengah
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_colwidth', 40)
print(f"Jumlah data CNN Indonesia: {len(df_cnn)}")
print("\n--- Preview Dataset CNN Indonesia (Struktur Awal) ---")

# Menampilkan dataset secara keseluruhan (otomatis muncul ... di tengah)
display(df_cnn[kolom_pilihan])
print("-" * 60)
print(f"Jumlah data Turnbackhoax: {len(df_tbh)}")
print("\n--- Preview Dataset Turnbackhoax (Struktur Awal) ---")

display(df_tbh[kolom_pilihan])
```

```python
# 2. Setting agar teks tampil panjang ke samping tanpa dipotong titik-titik di tengah
pd.set_option('display.max_colwidth', 500) 

# 3. Menampilkan 5 Sampel Data CNN Indonesia
print("--- SAMPEL 5 DATA MENTAH CNN INDONESIA ---")
display(df_cnn[['Timestamp', 'Title', 'FullText', 'Category']].head(5))

print("\n--- SAMPEL 5 DATA MENTAH TURNBACKHOAX ---")
display(df_tbh[['Timestamp', 'Title', 'FullText', 'Category']].head(5))

# 4. Kembalikan ke default agar output kodingan lain tidak terlalu lebar
pd.reset_option('display.max_colwidth')
```

# Pemberian Label

```python
# CNN Indonesia = 0 (Valid)
# Turnbackhoax = 1 (Hoaks)
df_cnn['label'] = 0
df_tbh['label'] = 1
```

```python
# 2. Melihat sampel data untuk memastikan label sudah terisi benar
print("\n--- Sampel 10 Data Teratas CNN (Valid - 0) ---")
display(df_cnn[['Title', 'FullText', 'label']].head(10))
```

```python
print("\n--- Sampel 10 Data Teratas Turnbackhoax (Hoaks - 1) ---")
display(df_tbh[['Title', 'FullText', 'label']].head(10))
```

# Penggabungan Atribut

```python
# Penggabungan Atribut (Title + FullText)
df_cnn['text_original'] = df_cnn['Title'].astype(str) + " " + df_cnn['FullText'].astype(str)
df_tbh['text_original'] = df_tbh['Title'].astype(str) + " " + df_tbh['FullText'].astype(str)
```

```python
# --- PENGATURAN VISUALISASI RAPI (10 BARIS) ---
pd.set_option('display.max_rows', 10)     # 5 atas, 5 bawah dengan ... di tengah
pd.set_option('display.max_colwidth', 40) # Lebar kolom dibatasi agar ringkas di dokumen

print(f"Total Data CNN: {len(df_cnn)} baris")
print("--- Preview Validasi Penggabungan Kolom (CNN Indonesia) ---")
display(df_cnn[['Title', 'FullText', 'text_original']])

print("-" * 85)

print(f"Total Data Turnbackhoax: {len(df_tbh)} baris")
print("--- Preview Validasi Penggabungan Kolom (Turnbackhoax) ---")
display(df_tbh[['Title', 'FullText', 'text_original']])
```

```python
# --- VALIDASI PENGGABUNGAN TEKS SECARA UTUH (CNN INDONESIA) ---

# 1. Matikan batasan lebar kolom agar teks terlihat semua tanpa dipotong (...)
pd.set_option('display.max_colwidth', None)

# 2. Pengaturan tampilan baris
pd.set_option('display.max_rows', 10)

print("--- Preview Hasil Penggabungan CNN Indonesia (Teks Utuh) ---")

# Menampilkan 5 data teratas (Awal)
display(df_cnn[['Title', 'FullText', 'text_original']].head(5))

print("-" * 30 + " Dan seterusnya (Data Tengah) " + "-" * 30)

# Menampilkan 5 data terbawah (Akhir)
display(df_cnn[['Title', 'FullText', 'text_original']].tail(5))
```

```python
# --- VALIDASI PENGGABUNGAN TEKS SECARA UTUH (TURNBACKHOAX) ---

# 1. Matikan batasan lebar kolom agar teks terlihat semua
pd.set_option('display.max_colwidth', None)

# 2. Tetap tampilkan 10 baris secara manual (5 atas, 5 bawah)
pd.set_option('display.max_rows', 10)

print("--- Preview Hasil Penggabungan Turnbackhoax (Teks Utuh) ---")

# Menampilkan 5 data teratas untuk melihat awal dataset
display(df_tbh[['Title', 'FullText', 'text_original']].head(5))

print("-" * 30 + " Dan seterusnya (Data Tengah) " + "-" * 30)

# Menampilkan 5 data terbawah untuk melihat akhir dataset
display(df_tbh[['Title', 'FullText', 'text_original']].tail(5))
```

# Penggabungan Dataset (Merging)

```python
columns_to_keep = ['Timestamp', 'text_original', 'label']
df_merged = pd.concat([df_cnn[columns_to_keep], df_tbh[columns_to_keep]], ignore_index=True)
```

```python
# 1. Melihat informasi dasar dataset yang sudah digabung
print("--- INFO DATASET TERGABUNG (df_merged) ---")
df_merged.info()
```

```python
# --- PREVIEW KESELURUHAN DATASET TERGABUNG ---

# 1. Hitung total sebelum di-drop (untuk laporan data mentah)
total_mentah = len(df_merged) 
print(f"--- Total Dataset Mentah: {total_mentah} Baris ---")

# 2. Proses pembersihan (yang bikin jumlahnya jadi 4.155)
df_merged.drop_duplicates(subset=['text_original'], keep='first', inplace=True)
total_bersih = len(df_merged)

print(f"--- Total Dataset Setelah Cleaning: {total_bersih} Baris ---")
print(f"--- (Dihapus {total_mentah - total_bersih} data duplikat) ---\n")

# 3. Pengaturan tampilan (5 atas, 5 bawah)
# max_rows 10 artinya pandas akan nampilin 5 awal dan 5 akhir kalau datanya banyak
pd.set_option('display.max_rows', 10) 
pd.set_option('display.max_colwidth', 100) # Biar teks original nggak kepanjangan

display(df_merged[['Timestamp', 'text_original', 'label']])

# Reset settingan display
pd.reset_option('display.max_rows')
```

```python
# 4. Mengecek distribusi label untuk memastikan tidak ada data yang hilang
print("\n--- Jumlah Data Per Label ---")
print(df_merged['label'].value_counts().rename({0: 'Valid (CNN)', 1: 'Hoax (TBH)'}))
```

# Data Cleaning Tingkat Awal (Rendudancy)

```python
df_merged.dropna(subset=['text_original'], inplace=True)
duplicate_count = df_merged.duplicated(subset=['text_original']).sum()
df_merged.drop_duplicates(subset=['text_original'], keep='first', inplace=True)
```

```python
# Mengecek apakah ada kolom yang kosong (null) setelah penggabungan
print("\n--- Pengecekan Missing Values ---")
print(df_merged.isnull().sum())
```

```python
# Identifikasi Data Redundan (Duplikat) Sebelum Pembersihan
# Menggabungkan data sementara untuk pengecekan kualitas dataset
df_temp = pd.concat([df_cnn[['Timestamp', 'text_original', 'label']], 
                     df_tbh[['Timestamp', 'text_original', 'label']]], ignore_index=True)

# Mengekstraksi seluruh baris yang memiliki kesamaan konten teks (redundansi)
df_tampil_duplikat = df_temp[df_temp.duplicated(subset=['text_original'], keep=False)]

print(f"--- Visualisasi Data Duplikat Terdeteksi (Total: {len(df_tampil_duplikat)} entri) ---")

# Menampilkan sampel data duplikat berdasarkan urutan teks untuk validasi kualitatif
display(df_tampil_duplikat.sort_values(by='text_original').head(8))
```

```python
print("\n--- Pembersihan Duplikat ---")
print(f"Jumlah data duplikat yang dihapus: {duplicate_count}")
print(f"Total data setelah pembersihan: {len(df_merged)}")
```

# Shuffling Data

```python
df_final = df_merged.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
print("\n--- Distribusi Label Akhir ---")
print(df_final['label'].value_counts().rename({0: 'CNN (Valid)', 1: 'TBH (Hoax)'}))
```

```python
# Menampilkan hasil
pd.set_option('display.max_rows', 10)

print(f"\n--- Visualisasi Struktur Dataset Akhir (Total: {len(df_final)} Baris) ---")
# Menampilkan struktur dataset dengan representasi baris awal dan akhir (ellipses)
display(df_final[['Timestamp', 'text_original', 'label']])
```

```python
# Preview hasil akhir data preparation
pd.set_option('display.max_rows', 10) 

print(f"--- Visualisasi Dataset (Total: {len(df_final)} Baris) ---")
display(df_final[['Timestamp', 'text_original', 'label']])
```

# Export Hasil

```python
output_file = 'dataset_merged_labeled.xlsx'
df_final.to_excel(output_file, index=False)

print(f"\n--- Proses File 1 Selesai ---")
print(f"Data tersimpan di: {output_file}")
```
